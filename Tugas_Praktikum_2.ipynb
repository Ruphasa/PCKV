5.	Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!

```python
random_rect_img = np.zeros((512, 512, 3), dtype=np.uint8)

for i in range(5):
    x1 = np.random.randint(0, 400)
    y1 = np.random.randint(0, 400)
    x2 = np.random.randint(x1 + 50, 512)
    y2 = np.random.randint(y1 + 50, 512)
    color = (np.random.randint(0, 256), np.random.randint(0, 256), np.random.randint(0, 256))
    thickness = np.random.randint(1, 10)
    cv.rectangle(random_rect_img, (x1, y1), (x2, y2), color, thickness)

plt.imshow(random_rect_img)
plt.title('Image with 5 Random Rectangles')
plt.show()
```

7.	Buat rectangle dan circle pada bagian wajah dari image foto anda saat beraktifitas (bukan pasfoto).

```python
img_face = cv.cvtColor(cv.imread('/content/drive/MyDrive/PCVK_MM/Images/myFace.jpg'), cv.COLOR_BGR2RGB)

rect_pt1 = (500, 600)
rect_pt2 = (800, 300)

circle_center = (650, 450)
circle_radius = 150

cv.rectangle(img_face, rect_pt1, rect_pt2, color=(255, 0, 0), thickness=10)

cv.circle(img_face, circle_center, circle_radius, color=(0, 255, 0), thickness=10)

plt.imshow(img_face)
plt.title('Image with Rectangle and Circle on Face')
plt.show()
```

8.	Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB dari citra kitten/ lena/ mandrill/ male/ female/ couple/ sailboat/ peppers!

```python
img_bgr = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/hanekawa_hitam.jpg')

b, g, r = cv.split(img_bgr)

blue_channel = b.copy()

height, width = blue_channel.shape

rect_pt1 = (0, height - 50)
rect_pt2 = (50, height)


cv.rectangle(blue_channel, rect_pt1, rect_pt2, color=(255), thickness=4)


plt.imshow(blue_channel, cmap='gray')
plt.title('Blue Channel with Rectangle in Bottom-Left Corner')
plt.show()
```

9.	Lengkapi tulisan nama file pada file citra dari soal no.8. gunakan font, ukuran font, dan warna font yang sesuai keinginan anda.

```python
font = cv.FONT_HERSHEY_SIMPLEX
text = 'hanekawa_hitam.jpg'
org = (10, 30)
fontScale = 0.8
color = (255)
thickness = 2
lineType = cv.LINE_AA

cv.putText(blue_channel, text, org, font, fontScale, color, thickness, lineType)

plt.imshow(blue_channel, cmap='gray')
plt.title('Blue Channel with Rectangle and Filename')
plt.show()
```
