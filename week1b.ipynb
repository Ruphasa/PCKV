## import library dan image yang ada di drive

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/hanekawa_hitam.jpg')
plt.imshow(img)
#perhatikan hasilnya adalah citra dgn channel warna BGR
```

## Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image dengan tipe data int16.

```python
black_img = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(black_img)
```

## Kemudian menambahkan bentuk persegi panjang sesuai koordinat pt1 dan pt2

```python
#perhatikan koordinat titik2 pt1 dan pt2
cv.rectangle(black_img, pt1=(384,0), pt2=(510,150), color=(0,255,0), thickness=10)
plt.imshow(black_img)
```

## Selanjutnya ditambah menambahkan bentuk persegi sesuai koordinat pt1 dan pt2 yang tertulis pada kode program.

```python
cv.rectangle(black_img, pt1=(200,200), pt2 = (300, 300), color= (0,0,255), thickness=15)
plt.imshow(black_img)
```

## Tahap selanjutnya ditambah menambahkan bentuk lingkaran sesuai radius yang tertulis pada kode program.

```python
cv.circle(black_img, center=(100, 100), radius=50, color=(255,0,0), thickness=8)
plt.imshow(black_img)
```

## Kemudian dilakukan penambahan garis sesuai koordinat pt1 dan pt2 sebagai berikut.

```python
cv. line (black_img, pt1=(0,0), pt2 = (512,512), color=(255, 255, 255), thickness=5)
plt.imshow(black_img)
```

## Penambahan text dengan font yang telah tertulis dengan ukuran yang sudah ditentukan.

```python
font=cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='Hello', org=(10,500), fontFace=font, fontScale=4, color=(255,255,0), thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)
```

5.	Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!

```python
# Create a black image
random_rect_img = np.zeros((512, 512, 3), dtype=np.uint8)

# Draw 5 random rectangles
for i in range(5):
    x1 = np.random.randint(0, 400)
    y1 = np.random.randint(0, 400)
    x2 = np.random.randint(x1 + 50, 512)
    y2 = np.random.randint(y1 + 50, 512)
    color = (np.random.randint(0, 256), np.random.randint(0, 256), np.random.randint(0, 256))
    thickness = np.random.randint(1, 10)
    cv.rectangle(random_rect_img, (x1, y1), (x2, y2), color, thickness)

plt.imshow(random_rect_img)
plt.title('Image with 5 Random Rectangles')
plt.show()
```

7.	Buat rectangle dan circle pada bagian wajah dari image foto anda saat beraktifitas (bukan pasfoto).

```python
img_face = cv.cvtColor(cv.imread('/content/drive/MyDrive/PCVK_MM/Images/myFace.jpg'), cv.COLOR_BGR2RGB)

# Define the coordinates for the rectangle and circle (adjust these for your image)
# Rectangle coordinates (top-left and bottom-right corners)
rect_pt1 = (500, 600)
rect_pt2 = (800, 300)

# Circle coordinates (center and radius)
circle_center = (650, 450)
circle_radius = 150

# Draw the rectangle
cv.rectangle(img_face, rect_pt1, rect_pt2, color=(255, 0, 0), thickness=10) # Red rectangle

# Draw the circle
cv.circle(img_face, circle_center, circle_radius, color=(0, 255, 0), thickness=10) # Green circle

# Display the image
plt.imshow(img_face)
plt.title('Image with Rectangle and Circle on Face')
plt.show()
```

8.	Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB dari citra kitten/ lena/ mandrill/ male/ female/ couple/ sailboat/ peppers!

```python
# Load an image (replace with the path to one of the specified images if needed)
# Make sure the image is in the current directory or provide the full path
img_bgr = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/hanekawa_hitam.jpg')

# Split the image into RGB channels
b, g, r = cv.split(img_bgr)

# Create a copy of the blue channel
blue_channel = b.copy()

# Get the dimensions of the blue channel
height, width = blue_channel.shape

# Define the coordinates for the rectangle in the bottom-left corner
# Adjust the size of the rectangle as needed
rect_pt1 = (0, height - 50)  # Top-left corner (x, y)
rect_pt2 = (50, height)    # Bottom-right corner (x, y)


# Draw the rectangle on the blue channel
cv.rectangle(blue_channel, rect_pt1, rect_pt2, color=(255), thickness=4) # Draw a filled white rectangle


# Display the blue channel with the rectangle
plt.imshow(blue_channel, cmap='gray')
plt.title('Blue Channel with Rectangle in Bottom-Left Corner')
plt.show()
```

9.	Lengkapi tulisan nama file pada file citra dari soal no.8. gunakan font, ukuran font, dan warna font yang sesuai keinginan anda.

```python
# Add text to the image
font = cv.FONT_HERSHEY_SIMPLEX
text = 'hanekawa_hitam.jpg'
org = (10, 30)  # Position of the text (bottom-left corner of the text string)
fontScale = 0.8
color = (255)  # White color for grayscale image
thickness = 2
lineType = cv.LINE_AA

cv.putText(blue_channel, text, org, font, fontScale, color, thickness, lineType)

# Display the image with the text
plt.imshow(blue_channel, cmap='gray')
plt.title('Blue Channel with Rectangle and Filename')
plt.show()
```
